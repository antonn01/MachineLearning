3) Ada beberapa macam taxonomi dalam penerapan machine learning yang mencakup:

     ##### Supervised Learning :
     
     * Supervised learning adalah jenis pembelajaran di mana model dilatih menggunakan data yang telah diberi label. Artinya, setiap contoh data memiliki label yang menunjukkan output yang diinginkan.
     * Tujuan dari supervised learning adalah untuk mempelajari hubungan antara input dan output sehingga model dapat membuat prediksi akurat terhadap data baru yang belum pernah dilihat sebelumnya.
     * Contoh-contoh algoritma supervised learning meliputi regresi linier, regresi logistik, pohon keputusan, dan jaringan saraf tiruan.
     * Aplikasi dari supervised learning meliputi prediksi harga saham, klasifikasi email spam, identifikasi wajah dalam gambar, dan banyak lagi.

     ##### Unsupervised Learning :
     * Unsupervised learning adalah jenis pembelajaran di mana model dilatih menggunakan data tanpa label. Artinya, model harus menemukan struktur atau pola dalam data tanpa bantuan label.
     * Tujuan dari unsupervised learning adalah untuk mengeksplorasi data dan menemukan pola tersembunyi yang mungkin tidak terlihat secara langsung.
     * Contoh-contoh algoritma unsupervised learning meliputi clustering (seperti K-means dan hierarchical clustering), asosiasi (seperti algoritma apriori), dan reduksi dimensi (seperti analisis faktor dan analisis komponen utama).
     * Aplikasi dari unsupervised learning meliputi segmentasi pasar, pengelompokan dokumen, pengurutan berita berdasarkan topik, dan lainnya.

     ##### Semi-supervised Learning :
     * Semi-supervised learning adalah kombinasi dari supervised dan unsupervised learning. Model dilatih menggunakan sejumlah kecil data berlabel dan sebagian besar data tanpa label.
     * Tujuan dari semi-supervised learning adalah untuk menggunakan informasi yang terbatas dari data berlabel untuk meningkatkan kinerja model pada data tanpa label.
     * Contoh-contoh aplikasi semi-supervised learning termasuk klasifikasi gambar dengan label terbatas, analisis sentimen dengan sedikit data yang diketik ulang, dan lainnya.

     ##### Reinforcement Learning :
     * Reinforcement learning adalah jenis pembelajaran di mana agen belajar bertindak di lingkungannya untuk mencapai tujuan tertentu. Agar agen dapat belajar, ia menerima umpan balik dalam bentuk reward atau punishment dari lingkungannya.
     * Tujuan dari reinforcement learning adalah untuk mempelajari kebijaksanaan atau kebijakan tindakan yang optimal untuk mencapai tujuan tertentu dalam lingkungan yang tidak selalu terduga.
     * Contoh-contoh aplikasi reinforcement learning meliputi pengendalian robot, permainan video, otomatisasi trading saham, dan lainnya.

````python
import pandas as pd

# Membuat DataFrame
data = {
    'HARI': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'DATANG': [2, 3, 4, 1, 2, 5, 2],
    'BIAYA': [30000, 35000, 25000, 15000, 20000, 30000, 35000],
    'MAHASISWA': ['Ani', 'Budi', 'Jono', 'Lono', 'Joni', 'Ani', 'Budi']
}

df = pd.DataFrame(data)

# Menghitung total biaya untuk setiap baris
df['TOTAL_BIAYA'] = df['DATANG'] * df['BIAYA']

# a) Rata-rata mahasiswa datang pada minggu ini
average_students = df['DATANG'].mean()
print(f"Rata-rata mahasiswa datang pada minggu ini: {average_students}")

# b) Hari dengan biaya tertinggi
highest_cost_day = df.loc[df['TOTAL_BIAYA'].idxmax()]
print(f"Hari dengan biaya tertinggi: {highest_cost_day['HARI']} dengan total biaya {highest_cost_day['TOTAL_BIAYA']}")

# c) Hari dengan biaya lebih dari 110000
days_high_cost = df[df['TOTAL_BIAYA'] > 110000]['HARI'].tolist()
print(f"Hari dengan biaya lebih dari 110000: {days_high_cost}")

# d) Mahasiswa yang paling banyak datang ke kampus
most_frequent_student = df.groupby('MAHASISWA')['DATANG'].sum().idxmax()
print(f"Mahasiswa yang paling banyak datang ke kampus: {most_frequent_student}")

# e) Mahasiswa yang datang pada hari Minggu
student_sunday = df[df['HARI'] == 'Minggu']['MAHASISWA'].tolist()
print(f"Mahasiswa yang datang pada hari Minggu: {student_sunday}")

# f) Biaya tertinggi dan terendah
highest_cost = df['TOTAL_BIAYA'].max()
lowest_cost = df['TOTAL_BIAYA'].min()
print(f"Biaya tertinggi: {highest_cost}")
print(f"Biaya terendah: {lowest_cost}")

# g) Frekuensi datang tertinggi dan terendah
highest_frequency = df['DATANG'].max()
lowest_frequency = df['DATANG'].min()
print(f"Frekuensi datang tertinggi: {highest_frequency}")
print(f"Frekuensi datang terendah: {lowest_frequency}")
````

````python
from plotnine import ggplot, aes, geom_bar
import pandas as pd

# Data
fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]

# Membuat DataFrame
info_mahasiswa = pd.DataFrame({
    'fakultas': fakultas,
    'jumlah_mahasiswa': jumlah_mahasiswa,
    'akreditasi': akreditasi
})

# Menampilkan DataFrame
print(info_mahasiswa)

# Membuat plot
gambar = ggplot(info_mahasiswa, aes(x='fakultas', y='jumlah_mahasiswa', fill='fakultas')) + geom_bar(stat='identity')
print(gambar)
````
